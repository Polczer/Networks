# Возможности программы Wireshark


## 1. Захват данных в реальном времени
Wireshark позволяет захватывать пакеты данных, проходящие через сетевой интерфейс, в реальном времени. Это полезно для:
- Анализа трафика сети.
- Диагностики сетевых проблем.
- Мониторинга производительности сети.
- Возможность сохранения отдельных пакетов или сегментов трафика.
        



## 2. Фильтрация пакетов
Wireshark предоставляет инструменты для фильтрации данных:
- Фильтры захвата для ограничения типов данных, которые записываются.
- Фильтры отображения для выбора определенных пакетов из захваченного трафика.





## 3. Расширенные возможности анализа
Wireshark позволяет:
- Просматривать детализированную информацию о каждом пакете.
- Использовать графики для анализа временных характеристик трафика.
- Сравнивать пакеты, чтобы находить отклонения в передаче данных.
- Множество других инструментов






Wireshark позволяет анализировать и диагностировать проблемы с сетью. С его помощью можно глубже понять, что происходит в сети и как исправить потенциальные проблемы.
